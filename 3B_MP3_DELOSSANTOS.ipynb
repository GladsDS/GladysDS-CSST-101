Part 1: Implement Basic Probability Calculations

1. Joint Probability, For two independent events A and B, the probability of both happening together is calculated by multiplying their probabilities: P(A and B) = P(A) × P(B)

2. Marginal Probability, To find the probability of a single event A, you sum up the probabilities of A occurring with all possible other events B: P(A) = Σ P(A and B)

3. Conditional Probability, Using Bayes' Theorem, the probability of A given that B has occurred is: P(A|B) = (P(B|A) × P(A)) / P(B)

Python Implementation, Here's a simple Python program that demonstrates calculating these probabilities:




```python
def joint_probability(p_a, p_b):
    return p_a * p_b

def marginal_probability(p_a, p_b_given_a):
    return p_a * p_b_given_a

def conditional_probability(p_b_given_a, p_a, p_b):
    return (p_b_given_a * p_a) / p_b

def main():
    # Input probabilities
    p_a = float(input("Enter probability of event A: "))
    p_b = float(input("Enter probability of event B: "))
    p_b_given_a = float(input("Enter probability of B given A: "))

    # Calculate and display probabilities
    print(f"Joint Probability: {joint_probability(p_a, p_b)}")
    print(f"Marginal Probability: {marginal_probability(p_a, p_b_given_a)}")
    print(f"Conditional Probability: {conditional_probability(p_b_given_a, p_a, p_b)}")

if __name__ == "__main__":
    main()
```

Part 2:

Bayesian Inference uses Bayes' Theorem to update the probability of a hypothesis, such as having a disease, when new evidence becomes available. The formula for Bayes' Theorem is:

In this formula,

P(Disease∣Positive Test) represents the probability of having the disease after receiving a positive test result, known as the posterior probability.

P(Positive Test∣Disease) is the likelihood, which indicates the probability of getting a positive test result if the disease is present.

P(Disease) is the prior probability, representing the initial chance of having the disease before any test results. Lastly,

P(Positive Test) is the marginal likelihood, which is the overall probability of getting a positive test result.

```python
def bayesian_inference(prior, likelihood, marginal):
    return (likelihood * prior) / marginal

def main():
    # Input probabilities
    prior = float(input("Enter prior probability of disease: "))
    likelihood = float(input("Enter likelihood of positive test given disease: "))
    marginal = float(input("Enter overall probability of positive test: "))

    # Calculate and display posterior probability
    print(f"Posterior Probability: {bayesian_inference(prior, likelihood, marginal)}")

if __name__ == "__main__":
    main()
```

Part 3:

Simulate Decision-Making Under Uncertainty We'll write a Python function to simulate trials based on success chances, rewards, and penalties, and calculate the average outcome.

```python
import random

def simulate_decision(prob_success, reward, penalty, trials):
    total_return = sum(reward if random.random() < prob_success else penalty for _ in range(trials))
    return total_return / trials

def main():
    # Input parameters
    prob_success = float(input("Enter probability of success: "))
    reward = float(input("Enter reward for success: "))
    penalty = float(input("Enter penalty for failure: "))
    trials = int(input("Enter number of trials: "))

    # Run simulation and display result
    average_return = simulate_decision(prob_success, reward, penalty, trials)
    print(f"Average Return: {average_return}")

if __name__ == "__main__":
    main()
```

Interactive Component

This program allows the user to input the probability of success, the reward amount, the penalty for failure, and the number of trials. It then runs the simulation and calculates the average return based on the inputs.

Part 4:

We will create a Python program that lets users choose between simulating a binomial distribution (like coin flips) or a normal distribution (like exam scores).

The program will generate the selected distribution and display a histogram based on the user’s input.

Python Implementation Here’s a Python program to achieve this:



```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_binomial(trials, prob_heads):
    return np.random.binomial(1, prob_heads, trials)

def simulate_normal(mean, std_dev, trials):
    return np.random.normal(mean, std_dev, trials)

def main():
    choice = int(input("Choose distribution (1: Binomial, 2: Normal): "))
    trials = int(input("Enter number of trials: "))

    if choice == 1:
        prob_heads = float(input("Enter probability of heads: "))
        data = simulate_binomial(trials, prob_heads)
        plt.hist(data, bins=2, edgecolor='black')
        plt.title('Binomial Distribution')
        plt.xlabel('Outcome')
    elif choice == 2:
        mean = float(input("Enter mean: "))
        std_dev = float(input("Enter standard deviation: "))
        data = simulate_normal(mean, std_dev, trials)
        plt.hist(data, bins=30, edgecolor='black')
        plt.title('Normal Distribution')
        plt.xlabel('Value')
    else:
        print("Invalid choice.")
        return

    plt.ylabel('Frequency')
    plt.show()

if __name__ == "__main__":
    main()
```

Interactive Component

This program prompts the user to choose between simulating a binomial distribution or a normal distribution. Based on the user’s input, it will allow you to simulate either a binomial or normal distribution and display the corresponding histogram.
